```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
dataset = pd.read_csv('titanic.csv')
print(dataset.head())

import matplotlib.pyplot as plt
import seaborn as sns

# Plot jumlah penumpang yang selamat dan tidak
sns.countplot(x='Survived', data=dataset)
plt.title('Distribusi Penumpang yang Selamat')
plt.show()

sns.countplot(x='Survived', hue='Sex', data=dataset)
plt.title('Keselamatan Berdasarkan Jenis Kelamin')
plt.show()


# Cek informasi dataset
print(dataset.info())

# Cek jumlah nilai kosong di setiap kolom
print(dataset.isnull().sum())

# Cek statistik ringkasan data numerik
print(dataset.describe())
```
